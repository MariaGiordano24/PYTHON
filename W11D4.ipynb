```python
#Esercizio
#Creiamo un DataFrame sintetico, che contiene i guadagni mensili di diverse annate, con il seguente codice: years = 5 guadagni = pd.DataFrame
#({"Mese": list("GFMAMGLASOND"*years), "Anno": np.repeat(list(range(years)), 12), "Valore": np.random.randint(800, 5000, 12*years)}) 
#• Calcola la somma cumulativa dei guadagni utilizzando il metodo .cumsum() • Come sopra, ma raggruppato per ogni anno usando prima un .groupby()
import pandas as pd
import numpy as np
years = 5
guadagni = pd.DataFrame({"Mese": list("GFMAMGLASOND"*years),
                                   "Anno": np.repeat(list(range(years)), 12), 
                                   "Valore": np.random.randint(800, 5000, 12*years)
                        })
guadagni["guadagni_cum"] = guadagni["Valore"].cumsum()
print(guadagni)
```

```python
#Come sopra, ma raggruppato per ogni anno usando prima un .groupby()
guadagni["cumulativo_anno"] = guadagni.groupby("Anno")["Valore"].cumsum()
print(guadagni)
```

```python
#Esercizio 1/2
#Dal database AdventureWorks estraiamo la tabella dimcustomer • Trasformiamo i nomi dei clienti in modo che abbiano solo lettere minuscole, e i
#cognomi in modo che abbiano solo lettere maiuscole • Sulla colonna EmailAddress, utilizzando il metodo .str.split(), estraiamo nome utente e dominio 
#• Sulla colonna Phone, estraiamo ogni parte del numero (ad es. da "1 (11) 500 555-0162" a ["1", "(11)", "500", "555-0162"]) 
#• Utilizzando il metodo .str.contains(), estraiamo tutti gli indirizzi e-mail che contengono il numero "21"
#Esercizio 2/2 
#• Estraiamo tutti gli indirizzi e-mail che contengono il numero "20" oppure il numero "10" • Calcolare la lunghezza di ogni indirizzo e-mail ed
#estrarre i cinque più lunghi e i cinque più corti • Modificare il dominio degli indirizzi e-mail da "adventure-works.com" a "aw-db.com" mediante il
#metodo .str.replace() • Dalla colonna AddressLine1 estraiamo tutti gli indirizzi che contengono la sottostringa "Street"
import pandas as pd
import numpy as np
query = "SELECT * FROM dimcustomer"
df= pd.read_sql(query, db_engine)
df
```

```python
#Trasformiamo i nomi dei clienti in modo che abbiano solo lettere minuscole, e i cognomi in modo che abbiano solo lettere maiuscole
df["FirstName"] = df["FirstName"].str.lower()
df["LastName"] = df["LastName"].str.upper()
display(df)
```

```python
df.columns
```

```python
df["EmailAddress"]
```

```python
#Sulla colonna EmailAddress, utilizzando il metodo .str.split(), estraiamo nome utente e dominio 
df['EmailAddress_split'] = df['EmailAddress'].str.split('@')
print(df["EmailAddress_split"])
```

```python
#Sulla colonna Phone, estraiamo ogni parte del numero (ad es. da "1 (11) 500 555-0162" a ["1", "(11)", "500", "555-0162"]) 
df["Phone"]
```

```python
df["Phone_split"] = df["Phone"].str.split(" ")
print(df["Phone_split"])
```

```python
#Utilizzando il metodo .str.contains(), estraiamo tutti gli indirizzi e-mail che contengono il numero "21"
mask_21 = df["EmailAddress"].str.contains("21", regex=False)
df_21 = df[mask_21]
print(df[mask_21]["EmailAddress"])        #Ho filtrato la colonna EmailAddress per estrarre solo le email che contengono il numero "21".
```

```python
#Estraiamo tutti gli indirizzi e-mail che contengono il numero "20" oppure il numero "10"
mask_20_10 = (df["EmailAddress"].str.contains("20", regex=False)) | (df["EmailAddress"].str.contains("10", regex=False))
df_20_10 = df[mask_20_10]
print(df_20_10["EmailAddress"])                      #Estraggo tutte le email che contengono il numero "20" oppure "10".
```

```python
#Calcolare la lunghezza di ogni indirizzo e-mail ed estrarre i cinque più lunghi e i cinque più corti
df["EmailLength"] = df["EmailAddress"].str.len()
print(df[["EmailAddress", "EmailLength"]])           #Ho calcolato la lunghezza di ogni indirizzo e-mail e la visualizziamo insieme all'indirizzo.
```

```python
#estrarre i cinque più lunghi e i cinque più corti
print(df.nlargest(5, "EmailLength")[["EmailAddress", "EmailLength"]])
print(df.nsmallest(5, "EmailLength")[["EmailAddress", "EmailLength"]])
```

```python
#Modificare il dominio degli indirizzi e-mail da "adventure-works.com" a "aw-db.com" mediante il metodo .str.replace()
df["EmailAddress"] = df["EmailAddress"].str.replace("adventure-works.com", "aw-db.com", regex=False)
print(df["EmailAddress"])
                                                #Sostituisco il dominio degli indirizzi e-mail da "adventure-works.com" a "aw-db.com" con .str.replace
```

```python
#Esercizio 1/2
#Dai beginner_datasets carichiamo in un DataFrame il file facebook.csv, che contiene dei post con data di pubblicazione, tipo (foto, video, …) e numero
#di reactions raccolte: • Con la funzione pd.to_datetime() convertiamo la colonna status_published in formato Timestamp • Utilizzando gli attributi .dt
#.year , .dt.month , .dt.day , .dt.dayofweek , .dt.dayofyear, ottieniamo informazioni specifiche sulle date delle transazioni, come l'anno, il mese, il
#giorno della settimana, il giorno dell'anno, eccetera
#Esercizio 2/2 
#• Estraiamo solo i post relativi al 2012 • Estraiamo solo i post relativi a maggio 2018 • Confrontiamo il numero di post pubblicati nei weekend
#rispetto al numero di post pubblicati nel resto della settimana • Troviamo il primo e ultimo post pubblicati in ogni anno • Quanti tipi di post ci
#sono? E quanti per ogni tipo?
import pandas as pd
df = pd.read_csv("facebook.csv")
df
```

```python
df["year"] = df["status_published"].dt.year
print(df["year"])                                     #Estraggo l'anno dalla colonna status_published usando l'attributo .dt.year
```

```python
df["month"] = df["status_published"].dt.month
print(df["month"])                                     #Estraggo il mese dalla colonna status_published usando l'attributo .dt.month
```

```python
df["day"] = df["status_published"].dt.day
print(df["day"])                                     #Estraggo il giorno dalla colonna status_published usando l'attributo .dt.day
```

```python
df["dayofweek"] = df["status_published"].dt.dayofweek
print(df["dayofweek"])                             #Estraggo il giorno della settimana dalla colonna status_published usando l'attributo .dt.dayofweek
```

```python
df["dayofyear"] = df["status_published"].dt.dayofyear
print(df["dayofyear"])                             #Estraggo il giorno dell'anno dalla colonna status_published usando l'attributo .dt.dayofyear
```

```python
post_2012 = df[df["year"] == 2012] 
display(post_2012)                           #Estraggo solo i post relativi al 2012
```

```python
post_maggio_2018 = df[(df["year"] == 2018) & (df["month"] == 5)]
display(post_maggio_2018)                                # Estraiamo solo i post relativi a maggio 2018
```

```python
weekend_mask = df["dayofweek"].isin([5, 6])  # Maschera per i post nel weekend
num_weekend_post = df[weekend_mask].shape[0]  # Numero di post nel weekend
num_rest_post = df[~weekend_mask].shape[0]  # Numero di post nel resto della settimana

print("Numero di post nel weekend:", num_weekend_post)
print("Numero di post nel resto della settimana:", num_rest_post)
```

```python
primo_post = df.groupby("year")["status_published"].min()    #Primo post pubblicato in ogni anno
ultimo_post = df.groupby("year")["status_published"].max()   # Ultimo post pubblicato in ogni anno
print("Primo post per anno:\n", primo_post)
print("Ultimo post per anno:\n", ultimo_post)
```
