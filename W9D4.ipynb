```python
#ESERCIZIO 1
#Abbiamo la stringa: nome_scuola = "Epicode" Stampare ogni carattere della stringa, uno su ogni riga, utilizzando un costrutto while.
nome_scuola = 'Epicode'
indice = 0                          #INIZIALIZZAZIONE
while indice < len(nome_scuola):
    print(nome_scuola[indice])
    indice += 1                     #AGGIORNAMENTO
```

```python
#Esercizio Stampare a video tutti i numeri da 0 a 20 utilizzando il costrutto while.
indice = 0                          #INIZIALIZZAZIONE
lunghezza = 20 
while indice <= lunghezza:
    print(indice)
    indice += 1                      #AGGIORNAMENTO

```

```python
#Esercizio Calcolare e stampare tutte le prime 10 potenze di 2 (e.g., 2⁰, 2¹, 2², …) utilizzando un ciclo while.
indice = 0                                      #INIZIALIZZAZIONE
lunghezza = 10
while indice <= lunghezza:
    print('2 ^', indice, '=', 2**indice)
    indice += 1                                 #AGGIORNAMENTO
print('Finito')
```

```python
#Esercizio Calcolare e stampare tutte le prime N potenze di 2 utilizzando un ciclo while, domandando all'utente di inserire N.
indice = 0                                    #INIZIALIZZAZIONE
N = 20
while indice <= N:
      print('2 ^', indice, '=', 2**indice)
      indice += 1

```

```python
#Esercizio Calcolare e stampare tutte le potenze di 2 minori di 25000
indice = 0 
potenza = 2** indice
while potenza < 25000:
      print('2 ^', indice, '=', potenza)
      indice += 1
      potenza = 2** indice
```

```python
#Esercizio Calcolare (ma non stampare) le prime N potenze di K; ognuna di esse andrà memorizzata in coda a una lista. Alla fine, stampare la lista
#risultante. Proviamo con diversi valori di K, oppure facciamola inserire all'utente. Realizzare due versioni:
#• con un ciclo while,
#• con un ciclo for.
N = 15
K = 3
for esponente in range(N):
    potenza = K** esponente 
    print(K, '^', esponente, '=', potenza)
    
    
```

```python
#Esercizio Abbiamo una lista con i guadagni degli ultimi 12 mesi: guadagni = [100, 90, 70, 40, 50, 80, 90, 120, 80, 20, 50, 50] usando un costrutto
#for, calcolare la media dei guadagni e stamparla a video.
guadagni = [100, 90, 70, 40, 50, 80, 90, 120, 80, 20, 50, 50]
somma = 0
for guadagno in guadagni:
    somma += guadagno
    print('somma =', somma, '-guadagno =', guadagno)
```

```python
#Esercizio Abbiamo una lista di parole: parole = ["Albergo", "Sedia", "Borgo", "Petalo", "Eremo", "Belvedere", "Semestre", "Esteta", "Sosta", 
#"Orpello", "Abete", "Orologio", "Cesta", "Ermellino"] stampiamo, per ogni parola, quante volte appare la lettera "e"; facciamo attenzione al fatto
#che appare sia maiuscola che minuscola.
parole = ["Albergo", "Sedia", "Borgo", "Petalo", "Eremo", "Belvedere", "Semestre", "Esteta", "Sosta", "Orpello", "Abete", "Orologio", 
         "Cesta", "Ermellino"]
for parola in parole :
    count_e = parola.lower().count('e')
    print('La parola ', parola, 'contiene', str(count_e), 'e')

```

```python
#Esercizio Abbiamo un dizionario che assegna ad ogni proprietario la sua auto: 
#dizionario_auto = {"Ada": "Punto", "Ben": "Multipla", "Charlie": "Golf", "Debbie": "107"} Con un ciclo, e usando il metodo .values(),
#stampiamo a video tutte le auto che non sono una Multipla.
dizionario_auto = {"Ada": "Punto", "Ben": "Multipla", "Charlie": "Golf", "Debbie": "107"}
for auto in dizionario_auto.values():
    if auto != "Multipla":
        print(auto)


```
