```python
#Esercizio Il dataset insurance.csv dei beginner_datasets contiene dati su caratteristiche e abitudini delle persone rispetto ai costi individuali per 
#le cure mediche come premio per le assicurazioni sulla salute. • Visualizziamone le dimensioni, un'anteprima, e osserviamo i nomi di colonna; 
#• Quali sono le medie di charges rispetto a region? Ci sono differenze significative? • E rispetto a smoker? E a sex? 
#• Quali sono i descrittori statistici di bmi? Quali sono minimo, media e massimo di charges rispetto ai diversi quartili dei valori di bmi?
import pandas as pd

df = pd.read_csv("insurance.csv")
df
```

```python
#Visualizziamone le dimensioni, un'anteprima, e osserviamo i nomi di colonna
print(df.shape)
print(df.head(5))
df.columns
```

```python
#Quali sono le medie di charges rispetto a region? Ci sono differenze significative?
media_region = df.groupby("region")["charges"].mean()     #Raggruppato per region,selezionato la colonna charges e calcolato la media.
print(media_region)
```

```python
#E rispetto a smoker? E a sex?
media_smoker = df.groupby("smoker")["charges"].mean()      #Raggruppato per smoker,selezionato la colonna charges e calcolato la media.
media_sex = df.groupby("sex")["charges"].mean()            #Raggruppato per sex,selezionato la colonna charges e calcolato la media.
print(media_smoker, media_sex)
```

```python
#Quali sono i descrittori statistici di bmi?
df["bmi"].describe()
```

```python
#Quali sono minimo, media e massimo di charges rispetto ai diversi quartili dei valori di bmi?
charges_quartili = df.groupby("bmi_quartile")["charges"].agg(["min", "mean", "max"])
print(charges_quartili)
```

```python
#Esercizio 
#Carichiamo in un DataFrame il dataset iris.csv dei beginner_datasets e calcoliamo:•La media della lunghezza dei petali di tutto il dataset
#•La media della lunghezza dei petali per ogni specie di Iris, utilizzando il metodo .groupby() • Media, minimo e massimo della larghezza dei sepali 
#per ogni specie, utilizzando .groupby() e .agg()
import pandas as pd

df = pd.read_csv("iris.csv")
df
```

```python
#La media della lunghezza dei petali di tutto il dataset.
lunghezza_med_petali = df["petal_length"].mean().round()     
print(lunghezza_med_petali)                                    #Creo la variabile lunghezza_med_petali, seleziono la colonna 'petal_length' e calcolo 
                                                               #la media arrotondata
```

```python
#La media della lunghezza dei petali per ogni specie di Iris, utilizzando il metodo .groupby().
media_ogni_specie = df.groupby("species")["petal_length"].mean()     #Raggruppo il dataset per specie e calcolo la media della lunghezza dei petali 
print(media_ogni_specie)                                             #di ciascun gruppo 
```

```python
#Media, minimo e massimo della larghezza dei sepali per ogni specie, utilizzando .groupby() e .agg().
statistiche_sepali = df.groupby("species")["sepal_width"].agg(['mean', 'min', 'max'])
print(statistiche_sepali)
                                     #Raggruppo il dataset per specie e calcolo la media, il minimo e il massimo della larghezza dei sepali 
                                     #utilizzando .agg()
```

```python
#Esercizio
#Nei beginner_datasets, il dataset wine.csv contiene delle analisi organolettiche su diverse qualità di vini:
#• Qual è la media di concentrazione alcolica per ogni qualità? Ci sono differenze? E rispetto alla media totale? 
#• C'è differenza nella concentrazione alcolica per vini bianchi e vini rossi? 
#• Rifacendo le analisi dei due punti precedenti ma per il pH, cambia qualcosa? • E per i solfati?
import pandas as pd

df = pd.read_csv("wine.csv")
df
```

```python
# Qual è la media di concentrazione alcolica per ogni qualità? Ci sono differenze? E rispetto alla media totale? 
media_alcol_per_qualita = df.groupby("quality")["alcohol"].mean()        #Calcolo la media della concentrazione alcolica per ogni qualità di vino e confronto con la media totale
media_totale = df["alcohol"].mean().round() 
print(media_alcol_per_qualita) 
print(media_totale)
```

```python
# C'è differenza nella concentrazione alcolica per vini bianchi e vini rossi? 
diff_alcol_vinib_vinir = df.groupby("type")["alcohol"].mean()             #Calcolo la media della concentrazione alcolica per i vini bianchi e rossi per confrontare eventuali differenze

print(diff_alcol_vinib_vinir)
```

```python
# Rifacendo le analisi dei due punti precedenti ma per il pH, cambia qualcosa? 
pH = df.groupby("quality")["pH"].mean() 
print(pH)
```

```python
# E per i solfati?
pH = df.groupby("type")["pH"].mean()                   #Calcolo la media del pH per ogni qualità di vino per confrontare eventuali differenze tra le qualità

print(pH)
```

```python
#Esercizio 1/3
#Nei beginner_datasets, il file boston.csv contiene il Boston Housing Dataset, che deriva dalle informazioni raccolte dal Census Service degli Stati
#Uniti sulle abitazioni nell'area di Boston. Di seguito vengono descritte le colonne del dataset:
#• CRIM -tasso di criminalità pro capite per città 
#• ZN - proporzione di terreni residenziali suddivisi in zone per lotti superiori a 25.000 piedi quadrati. 
#• INDUS - percentuale di acri di attività commerciali non al dettaglio per città.
#• CHAS -variabile dummy del fiume Charles (1 se il tratto costeggia il fiume; 0 altrimenti)
#Esercizio 2/3 
#• NOX -concentrazione di ossidi di azoto (parti per 10 milioni). 
#• RM -numero medio di stanze per abitazione 
#• AGE -proporzione di unità abitative occupate da proprietari costruite prima del 1940 
#• DIS - distanze ponderate da cinque centri occupazionali di Boston
#• RAD -indice di accessibilità alle autostrade radiali 
#• TAX -aliquota dell'imposta fondiaria sul valore pieno per 10.000 dollari
#• PTRATIO - rapporto alunni-insegnanti per città
#• BLACK -la percentuale di neri per città 
#• LSTAT - % di popolazione di condizione più bassa
#Esercizio 3/3 
#• La media del prezzo delle case cambia a seconda della distanza dal fiume Charles? 
#• Si nota una correlazione tra il tasso di criminalità e il valore delle abitazioni? Come si può spiegare il risultato? 
#• Qual è la media del numero di stanze rispetto al rapporto alunni-insegnanti? E del valore delle case? Appare esserci una qualche correlazione?
#Come si può spiegare il risultato? 
#• Rispetto all'accessibilità alle autostrade, cambia qualcosa la media delle età delle abitazioni? E del numero di stanze? E delle tasse?
import pandas as pd

df = pd.read_csv("boston.csv")
df
```

```python
df.info()
```

```python
#CHAS -variabile dummy del fiume Charles (1 se il tratto costeggia il fiume; 0 altrimenti)
media_case_fiume = df.groupby("chas")["medv"].mean()   #Confronto il valore medio delle case in base al fiume Charles (chas = 0 → non costeggia,
print(media_case_fiume)                                #chas = 1 → costeggia)
```

```python
#La media del prezzo delle case cambia a seconda della distanza dal fiume Charles? 
media_case_per_fiume = df.groupby("chas")["medv"].mean()      # Confronto la media del prezzo delle case a seconda della vicinanza al fiume Charles
print(media_case_per_fiume)
```

```python
#Si nota una correlazione tra il tasso di criminalità e il valore delle abitazioni? Come si può spiegare il risultato?
correlazione_crim_medv = df["crim"].corr(df["medv"].round())    #Calcolo la correlazione tra tasso di criminalità e valore delle abitazioni e poi arrotondo 
print(correlazione_crim_medv)                                   #arrotondo il risultato
```

```python
#Qual è la media del numero di stanze rispetto al rapporto alunni-insegnanti?     
media_stanze_ptratio = df.groupby("ptratio")["rm"].mean().round(2).reset_index()
media_stanze_ptratio.columns = ["PTRATIO", "Media_RM"]
print(media_stanze_ptratio)                                          #Calcolo la media del numero di stanze in base al rapporto alunni-insegnanti
 
```

```python
#del valore delle case? Appare esserci una qualche correlazione? Come si può spiegare il risultato? 
media_valore_ptratio = df.groupby("ptratio")["medv"].mean().round(2).reset_index()
media_valore_ptratio.columns = ["PTRATIO", "Media_MEDV"]
print(media_valore_ptratio)                              #Calcolo la media del valore delle case in base al rapporto alunni-insegnanti
```

```python
#Rispetto all'accessibilità alle autostrade, cambia qualcosa la media delle età delle abitazioni? E del numero di stanze? E delle tasse?
media_autostrade = df.groupby("rad")[["age", "rm", "tax"]].mean().round(2).reset_index()
print(media_autostrade)                   #Analisi dell'influenza dell'accessibilità alle autostrade (RAD) su età delle abitazioni(age), 
                                          #numero di stanze(rm) e tasse (tax).
```

```python
#Esercizio 1/3
#Abbiamo un DataFrame di dipendenti: employees_df = pd.DataFrame({ 'employee_id': [101, 102, 103, 104, 105], 
#'name': ['Alice', 'Bob', 'Charlie', 'David', 'Emma'],
#'department_id': [1, 2, 1, 2, 3] })
#Esercizio 2/3 
#E un DataFrame di dipartimenti: departments_df = pd.DataFrame({ 'department_id': [1, 2, 3],
#'department_name': ['HR', 'IT', 'Finance'], 'location': ['New York', 'San Francisco', 'Chicago'] })
#Esercizio 3/3
#• Unire questi DataFrame in base alla colonna comune department_id, in modo da avere nel risultato informazioni sia sui dipendenti 
#che sui dipartimenti, usando la funzione  • Per ogni DataFrame, trasformare la colonna department_id nell'indice,
#facendo in modo che la modifica sia permanente; poi unire i due dataset mediante il metodo .join() 
#• Ci sono differenze nel risultato? Quali? Perché?

import pandas as pd

employees_df = pd.DataFrame({
    'employee_id': [101, 102, 103, 104, 105],
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Emma'],
    'department_id': [1, 2, 1, 2, 3]
})

departments_df = pd.DataFrame({
    'department_id': [1, 2, 3],
    'department_name': ['HR', 'IT', 'Finance'],
    'location': ['New York', 'San Francisco', 'Chicago']
})

#.merge()
employees_departments_merge = employees_df.merge(departments_df, on="department_id")
print(employees_departments_merge)
```

```python
#trasformare la colonna department_id nell'indice facendo in modo che la modifica sia permanente
employees_df.set_index("department_id", inplace=True)     
departments_df.set_index("department_id", inplace=True)
```

```python
employees_departments_joined = employees_df.join(departments_df)
print(employees_departments_joined)
```
